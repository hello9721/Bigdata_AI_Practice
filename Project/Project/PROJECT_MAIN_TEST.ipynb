```python
## DB 연결

con = sql.connect( host = '127.0.0.1', user = 'root', password = 'bigdatar', db = 'bigdata_temp', charset = 'utf8')
cursor = con.cursor()
```

```python
## DB 에서 데이터 추출

query = 'SELECT * FROM ORG_DATA ORDER BY S_TIME DESC LIMIT 1000'

cursor.execute(query)
result = cursor.fetchall()
nrow = cursor.rowcount

df = [[0] * 7 for i in range(nrow)]

for i in range(nrow):
    
    for j in range(7): df[i][j] = result[i][j]
        
col_1 = [i[1] for i in df]
col_2 = [i[2] for i in df]
col_3 = [i[3] for i in df]
col_4 = [i[4] for i in df]
col_5 = [i[5] for i in df]
col_6 = [i[6] for i in df]
```

```python
## 그래프 그려보기

pt.plot(col_1[ :150], label = 'TEMP 1')
pt.plot(col_2[ :150], label = 'TEMP 2')
pt.plot(col_3[ :150], label = 'TEMP 3')
pt.plot(col_4[ :150], label = 'TEMP 4')
pt.plot(col_5[ :150], label = 'TEMP 5')
pt.plot(col_6[ :150], label = 'TEMP 6')

pt.show()
```
